```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from mlxtend.plotting import plot_decision_regions
```

```python
np.random.seed(1)
X_xor = np.random.randn(500, 2)
y_xor = np.logical_xor(X_xor[:, 0] > 0, X_xor[:, 1] > 0)
y_xor = np.where(y_xor, 1, -1)

np.random.seed(100)
X_test = np.random.randn(500, 2)
y_test = np.logical_xor(X_test[:, 0] > 0, X_test[:, 1] > 0)
y_test = np.where(y_test, 1, -1)
```

```python
plt.scatter(X_xor[y_xor == 1, 0], X_xor[y_xor == 1, 1], 
            c='darkorange', marker='^', s=35, label='1', 
            edgecolors='black', alpha=0.8)

plt.scatter(X_xor[y_xor == -1, 0], X_xor[y_xor == -1, 1],
            c='steelblue', marker='s', s=35, label='-1', 
            edgecolors='black', alpha=0.8)

plt.tight_layout()
plt.legend()
plt.show()
```

# Support vector machine (kernel trick)

```python
svm = SVC(kernel='rbf', gamma=0.3, C=100)
svm.fit(X_xor, y_xor)

plot_decision_regions(X_xor, y_xor, clf=svm)
plt.title("Score: " + str(svm.score(X_test, y_test)))
plt.show()
```

# Decision tree

```python
from sklearn.tree import plot_tree

dtc = DecisionTreeClassifier(max_depth=6, criterion="gini", splitter="best")
dtc.fit(X_xor, y_xor)

plot_decision_regions(X_xor, y_xor, clf=dtc)
plt.title("Score: " + str(dtc.score(X_test, y_test)))
plt.show()

plot_tree(dtc)
plt.show()
```

# K-nearest Neighbor

```python
krn = KNeighborsClassifier()
krn.fit(X_xor, y_xor)

plot_decision_regions(X_xor, y_xor, clf=krn)
plt.title("Score: " + str(krn.score(X_test, y_test)))
plt.show()
```
